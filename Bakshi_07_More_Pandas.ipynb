```python
import pandas as pd

# loading the data
data = pd.read_csv('/content/ride_sharing.csv')
```

```python
# counts how many rows are in the data set
rows = str(len(data))
# counts how many columns are in the data set
columns = str(len(data.columns))

print('The number of rows are ' + rows + '.')
print('Thhe numbers columns are ' + columns + '.')
```

```python
#data.dtypes # checking the data type of the columns
#data.head(10) # visualizing the type of data in the duration column
data['duration'] = data['duration'].astype(str).str.extract('(\d+)') # converts the objects into strings and extracts the integer part of the string
data['duration'] = data['duration'].astype(int) # converts the extracted strings into integers
data.dtypes # displays the data types in the updated data frame
```

```python
min_value = data['duration'].min() # finding the minimum value and defining it as min_value
max_value = data['duration'].max() # finding the maximum value and defining it as max_value
print('Minimum value:', min_value)
print('Maximum value:', max_value)
```

```python
data_copy = data.copy() # making a copy of the origional data to modify
rides = data_copy.query('duration <= 60') # filtering the data so only rows with durations less than of equal to 60 are included
rides.head(10) # checking the new data frame
```

```python
min_value = rides['duration'].min() # finding the minimum value of the new dataset and defining it as min_value
max_value = rides['duration'].max() # finding the maximum value of the new dataset and defining it as max_value
print('Minimum value:', min_value)
print('Maximum value:', max_value)
```

```python
tire_size_26 = rides.query('tire_size == 26') # collects only the rows were the column tire_size is 26
tire_size_26.head(5) # checking the new data frame
```

```python
female_riders = tire_size_26.query("user_gender == 'Female'") # collects only the rows with female riders
#female_riders.head(10) # checking the new data frame
num_rows = len(female_riders) # counts the rows with famle riders
print('Number of female riders using the 26-inch wheeled rental bikes for one hour or less is ' + str(num_rows) + '.')

```

```python
messy_data = pd.read_csv('/content/more-messy-data.csv') # uploading the data
messy_data.head(5) # visualizing the data frame
```

```python
print('The number of duplicates rows are ' + (str(messy_data.duplicated().sum())) + '.') # shows the number of duplicate rows in the data frame
```

```python
data_copy1 = messy_data.copy() # makes a copy of the messy data
cleaned_data = data_copy1.drop_duplicates() # drops all of the duplicated in the
print(cleaned_data)
print('Now the number of duplicated rows are ' + str((cleaned_data.duplicated().sum())) + '.') # couting the duplicates as a method to douple check duplicates
```

```python
columns_to_fill = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width'] # specifies the four columns involving sepal and petal length and width
filled_data = cleaned_data.fillna(cleaned_data[columns_to_fill].mean(numeric_only=True)) # updates all NaN values with the selected columns means, considering only numeric values
print(filled_data) # printed new data frame to double check fot NaN values
```

```python
without_missing_data = filled_data.dropna() # drops any missing data from the cleaned and filled data frame
```

```python
species_count = without_missing_data['species'].nunique() # counts how many unique values there are under the species column
print('The number of species are ' + str(species_count) + '.')
unique_species = without_missing_data['species'].value_counts() # displays all the different species to help identify variations
print(unique_species)
```

```python
without_missing_data_copy = without_missing_data.copy() # makes a copy of the data set to modify incase origional needs to be acessable
without_missing_data_copy['species'] = without_missing_data_copy['species'].str.lower() # lowercases all of the values under the species column
typo_mapping = { # changes the common typos within the species column into the correct species value
    'setoso': 'setosa',
    'virginia': 'virginica',
    'viginica': 'virginica',
    'setosan' : 'setosa'
}
without_missing_data_copy['species'] = without_missing_data_copy['species'].replace(typo_mapping)
species_count = without_missing_data_copy['species'].nunique() # counts how many unique values there are under the species column in the new, cleaned copy
print('The number of species are now ' + str(species_count) + '.')

unique_species = without_missing_data_copy['species'].value_counts() # displays the different species in the cleaned data
print(unique_species)

```

```python
import re
# [A-Z]{3} means 3 uppercase letters in the alphabet in a row
# [\d]{4} means 4 numners 0-9 in a row
expression = '[A-Z]{3}[\d]{4}'

# list of strings to test
tester_strings = ['ABC1234', 'a1b23c4', 'ABCD1234', 'aBc123']

# function to test whether the string matched the expressiosn general format
def match(plate):
  if re.match(expression, plate):
    print("match!")
  else:
    print("no match!")
for string in tester_strings:
  match(string)
```
